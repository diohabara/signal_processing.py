# 4. ギブス現象と総和法

## 章末問題

```python
# 4.30
import numpy as np
import matplotlib.pyplot as plt

def partialsumsaw(t, n):
    ps = 0
    for k in range(1, n+1):
        ps += 2 * (-1)**(k-1) * np.sin(k*t) / k
    return ps

PI = np.pi
def saw(t):
    return t

npsaw = np.vectorize(saw)
t = np.linspace(-PI, PI, 10000)
x = npsaw(t)
y = partialsumsaw(t, 5)

plt.plot(t, x)
plt.plot(t, y)
plt.show()
```

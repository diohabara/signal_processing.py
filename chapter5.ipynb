# 5. 複素フーリエ級数

## サンプルコード

```python
# 5.1
import numpy as np
import matplotlib.pyplot as plt

m = 50
PI = np.pi
t = np.linspace(-PI, PI, 10000)
x = np.cos(t) * np.cos(m * t)
plt.plot(t, x)
plt.show()
```

```python
# 5.37
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-10, 10, 10000)
x = np.exp(np.cos(t)) * np.cos(np.sin(t))
plt.plot(t, x)
plt.show()
```

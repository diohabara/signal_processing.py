## サンプルコード

```python
# 1.1
import numpy as np

A = np.array([[1, 2, 3], [-1, 3, 5]])
B = np.array([[-2, 3], [1, -1], [2, -3]])
u = np.array([1, 2, 5]).reshape(3, 1)

C = np.dot(A, B)
v = np.dot(A, u)
print(C)
print(v)
```

```python
# 1.2
import numpy as np

def stepfun(t):
    if t >= 5:
        return 1
    elif t >= 0:
        return 0
    else:
        return -1

npstepfun = np.vectorize(stepfun)

t = np.linspace(-5, 10, 16)
print(npstepfun(t))
```

```python
# 1.3
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(-3, 3, 1000)
plt.plot(t, np.exp(-t**2) * np.sin(20 * t))
plt.title("sample graph $x(t)=e^{-t-2}\sin(20t)$")
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
# 1.4
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
x1 = np.sin(t ** 2)
x2 = 2 * np.sin(t ** 2) * np.exp(-t ** 2 / 20)
x3 = np.sin(t) ** 3
x4 = np.sqrt(np.abs(t)) * np.sin(t)
x5 = np.cos(t)

fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(t, x1)
ax1.set_ylim(-2.5, 2.5)

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(t, x2)
ax2.set_ylim(-2.5, 2.5)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(t, x3)
ax3.plot(t, x5)
ax3.set_ylim(-2.5, 2.5)

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(t, x4)
ax4.set_ylim(-2.5, 2.5)

plt.show()
```

```python
# 1.7
t = np.linspace(-2, 5, 1000)
def stepfun(t):
    if t >= 1:
        return np.exp(-(t - 2) ** 2)
    elif 0 <= t < 1:
        return 1
    else:
        return np.cos(10 * t)
npstepfun = np.vectorize(stepfun)

plt.plot(t, npstepfun(t))
plt.xlabel("t")
plt.ylabel("$f(t)$")
plt.show()
```

```python
# 1.8
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(-3, 3, 1000)
# pyplot
plt.plot(t, np.exp(-t**2) * np.sin(20 * t), color = "red", linewidth = 3.0, linestyle = "--")
plt.title("sample graph $x(t)=e^{-t-2}\sin(20t)$")
plt.xlabel("t")
plt.ylabel("x")
plt.show()

# object
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(t, np.exp(-t**2) * np.sin(20 * t), color = "red", linewidth = 3.0, linestyle = "--")
ax1.set_ylim(-1, 1)
```

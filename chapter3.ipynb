# 3. 関数の直交性

## サンプルコード

```python
# 3.1
import sympy as sym

PI = sym.S.Pi

t = sym.Symbol("t")
I = sym.integrate((sym.sin(t) - sym.cos(t))**2, (t, -PI, PI))
print(I)
```

```python
# 3.2
import sympy
t = sympy.Symbol("t")

for j in range(6):
    print(sympy.legendre(j, t))
```

```python
# 3.3
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import eval_legendre

t = np.linspace(-1, 1, 256)
legen = eval_legendre(11, t)
plt.plot(t, legen)
plt.show()
```

## 章末問題

```python
# 3.23
import sympy as sym
PI = sym.S.Pi
t = sym.Symbol("t")
I = sym.integrate((sym.sin(t) - sym.cos(t)**3)**2, (t, -PI, PI))
J = sym.integrate(sym.cos(t)**6, (t, -PI, PI))
print(I)
print(J)
```

```python
# 3.29
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import eval_chebyc

t = np.linspace(-1, 1, 256)

C = eval_chebyc(11, t)
plt.plot(t, C)
plt.show()
```
